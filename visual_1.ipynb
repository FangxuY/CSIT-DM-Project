```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datafile = pd.read_csv('data21_2.csv', low_memory = False)
```

```python
nonNan = []
#print(datafile['HOSPDAYS'].isnull())
for i, values in enumerate(datafile['HOSPDAYS'].isnull()):
    if values != True:
        nonNan.append(datafile['HOSPDAYS'][i])
#print(nonNan)
freq = {}
for i in nonNan:
    if freq == {}:
        freq[i] = nonNan.count(i)
    if i not in freq.keys():
        freq[i] = nonNan.count(i)
#print(freq)
#days = freq.keys()
#frequence = freq.values()
freqs = {}
#print(days)
#print(frequence)
for i in sorted(freq.keys()):
    freqs[i] = freq[i]
print(freqs)
x = list(freqs.keys())
y = list(freqs.values())
plt.figure(figsize = (30, 10))
plt.bar(x, y)
plt.plot(x, y, color = 'red')
plt.title("Hospital days vs frequence")
plt.xlabel("Hospital days")
plt.ylabel("Frequence")
plt.xticks(list(freqs.keys()))
plt.show()
#freq = {}
#freq[1] = 'a'
#freq[2] = 'b'
#print(freq)
#print(1 not in freq.keys())
```
