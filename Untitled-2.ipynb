```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
datafile = pd.read_csv('data21_2.csv', low_memory=False)

hospdays = datafile['HOSPDAYS'].dropna().astype('int')
# freqs = hospdays.value_counts().sort_index()
freqs = hospdays[hospdays <= 30].value_counts().sort_index()
freqs
```

```python
x = list(freqs.index)
y = list(freqs.values)
plt.figure(figsize=(8, 4), dpi=150)
plt.bar(x, y)
plt.plot(x, y, color='red')
plt.title('Hospital days vs Frequency')
plt.xlabel('Hospital days')
plt.ylabel('Frequence')
plt.show()
```

```python
symptoms = pd.concat([datafile['SYMPTOM1'], datafile['SYMPTOM2'], datafile['SYMPTOM3'], datafile['SYMPTOM4'], datafile['SYMPTOM5']]).dropna().value_counts()
symptoms[:15]
```

```python
top_symptoms = symptoms[:15].sort_values(ascending=True)
top_symptoms
```

```python
x = list(top_symptoms.index)
y = list(top_symptoms.values)
plt.figure(figsize=(8, 4), dpi=100)
plt.barh(x, y)
plt.show()
```
